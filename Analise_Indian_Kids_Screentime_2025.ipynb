```python
# importa√ß√µes

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Configura√ß√£o do seaborn
sns.set(style="whitegrid")

# Carregar os dados

df = pd.read_csv("Indian_Kids_Screen_Time.csv")

# Tipos de dados das colunas
tipos_dados = df.dtypes
```

```python
# tratando titulos
df.columns = [
    "Idade",
    "G√™nero",
    "Tempo M√©dio de Tela (h/dia)",
    "Dispositivo Principal",
    "Excedeu Limite Recomendado",
    "Rela√ß√£o Educacional/Recreativo",
    "Impactos na Sa√∫de",
    "√Årea (Urbana ou Rural)"
]

# --- Coluna saude ---

traducao = {
    "Poor Sleep": "Sono Ruim",
    "Eye Strain": "Cansa√ßo Visual",
    "Anxiety": "Ansiedade",
    "Obesity Risk": "Risco de Obesidade"
}

df['Impactos na Sa√∫de'] = df['Impactos na Sa√∫de'].dropna().apply(
    lambda x: ', '.join([traducao.get(i.strip(), i) for i in x.split(',')])
)
```

```python
print(df.columns)
print(df.head())
```

```python
# Medidas de tend√™ncia central
print("\nM√©dia:\n", df[['Idade', 'Tempo M√©dio de Tela (h/dia)', 'Rela√ß√£o Educacional/Recreativo']].mean())
print("\nMediana:\n", df[['Idade', 'Tempo M√©dio de Tela (h/dia)', 'Rela√ß√£o Educacional/Recreativo']].median())
print("\nModa:\n", df.mode().iloc[0][['Idade', 'Tempo M√©dio de Tela (h/dia)', 'Rela√ß√£o Educacional/Recreativo']])

# Medidas de dispers√£o
print("\nVari√¢ncia:\n", df[['Idade', 'Tempo M√©dio de Tela (h/dia)', 'Rela√ß√£o Educacional/Recreativo']].var())
print("\nDesvio Padr√£o:\n", df[['Idade', 'Tempo M√©dio de Tela (h/dia)', 'Rela√ß√£o Educacional/Recreativo']].std())
print("\nAmplitude:\n", df[['Idade', 'Tempo M√©dio de Tela (h/dia)', 'Rela√ß√£o Educacional/Recreativo']].max() - df[['Idade', 'Tempo M√©dio de Tela (h/dia)', 'Rela√ß√£o Educacional/Recreativo']].min())
```

```python
# Histograma distribui√ß√£o de tempo de tela
plt.style.use("seaborn-v0_8")
plt.figure(figsize=(9,5),facecolor='#e8eef1')
plt.hist(df['Tempo M√©dio de Tela (h/dia)'], bins=15, color='steelblue', edgecolor='white', alpha=0.9)

plt.title('Distribui√ß√£o do tempo de tela di√°rio (horas)', fontsize=16, fontweight='bold')
plt.xlabel('Horas por dia', fontsize=13)
plt.ylabel('N√∫mero de crian√ßas', fontsize=13)

plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.box(False)
plt.show()
```

```python
# Histograma: m√©dia de tempo de tela por √°rea urbana/rural
plt.style.use("seaborn-v0_8")
plt.figure(figsize=(9,5), facecolor='#e8eef1')

plt.hist(df[df['√Årea (Urbana ou Rural)']=='Urban']['Tempo M√©dio de Tela (h/dia)'],
         bins=12, color='steelblue', edgecolor='white', alpha=0.7, label='Urbano')
plt.hist(df[df['√Årea (Urbana ou Rural)']=='Rural']['Tempo M√©dio de Tela (h/dia)'],
         bins=12, color='lightgreen', edgecolor='white', alpha=0.7, label='Rural')

plt.title('Distribui√ß√£o de tempo de tela por √°rea', fontsize=16, fontweight='bold')
plt.xlabel('Horas por dia', fontsize=13)
plt.ylabel('N√∫mero de crian√ßas', fontsize=13)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.legend()
plt.box(False)

plt.show()
```

```python
# Gr√°fico de dispers√£o: rela√ß√£o entre idade e tempo de tela
plt.style.use("seaborn-v0_8")
plt.figure(figsize=(11,4),facecolor='#e8eef1')

sns.scatterplot(
    x='Tempo M√©dio de Tela (h/dia)',
    y='Idade',
    data=df,
    color='steelblue',
    s=60,
    alpha=0.8
)

plt.title('Dispers√£o de idade por tempo de tela di√°rio', fontsize=16, fontweight='bold')
plt.xlabel('Horas de tela por dia', fontsize=13)
plt.ylabel('Idade', fontsize=13)

max_horas = int(df['Tempo M√©dio de Tela (h/dia)'].max()) + 1
plt.xticks(range(0, max_horas+1, 1))

min_idade = int(df['Idade'].min())
max_idade = int(df['Idade'].max()) + 1
plt.yticks(range(min_idade, max_idade, 1))
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.box(False)

plt.show()
```

```python
# Gr√°fico de pizza: propor√ß√£o que excede o limite recomendado

limite = 2
excede = df['Tempo M√©dio de Tela (h/dia)'] > limite
counts = excede.value_counts()
counts = counts.reindex([False, True])

plt.style.use("seaborn-v0_8")
fig, ax = plt.subplots(figsize=(5,5), facecolor='#e8eef1')

cores = ['#4C72B0', '#DD614A']

wedges, texts, autotexts = ax.pie(
    counts,
    autopct='%1.1f%%',
    colors=cores,
    startangle=90,
    wedgeprops={'edgecolor':'white'}
)

for autotext in autotexts:
    autotext.set_fontsize(12)
    autotext.set_color('white')
    autotext.set_fontweight('bold')

ax.legend(
    wedges,
    ['Dentro do limite', 'Excede o limite'],
    title="Legenda",
    loc="center left",
    bbox_to_anchor=(1, 0.5),
    fontsize=12,
    title_fontsize=13,
    frameon=True,
    facecolor='#f5f5f5'
)

fig.suptitle(
    f'Propor√ß√£o de crian√ßas que excedem {limite}h de tela por dia',
    fontsize=16,
    fontweight='bold',
    x=0.67,
    ha='center',
    y=0.95
)

plt.show()
```

```python
# @title
# Grafico de barras: Principais impactos na sa√∫de
impacts = df['Impactos na Sa√∫de'].dropna().str.split(', ').explode()
impacts = impacts.str.strip()
impacts = impacts.replace({
    "Poor Sleep": "Sono Ruim",
    "Eye Strain": "Cansa√ßo Visual",
    "Anxiety": "Ansiedade",
    "Obesity Risk": "Risco de Obesidade"
})

plt.style.use("seaborn-v0_8")
plt.figure(figsize=(8,5), facecolor='#e8eef1')
ax = plt.gca()
ax.set_facecolor('#e8eef1')

impacts.value_counts().plot(
    kind='bar',
    color='steelblue',
    edgecolor='white',
    alpha=0.8
)
plt.title('Principais impactos na sa√∫de', fontsize=16, fontweight='bold')
plt.ylabel('N√∫mero de ocorr√™ncias', fontsize=13)
plt.xlabel('Impactos na sa√∫de', fontsize=13)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.box(False)
plt.show()
```
